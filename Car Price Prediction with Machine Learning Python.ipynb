```python
import pandas as pd 
data = pd.read_csv('cars.csv')
data
```

```python
data.drop(columns=['Car_ID'], inplace=True)
data
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
values = data['Year'].value_counts()
values.plot(kind='bar')
plt.show()
```

```python
data['Year'] = data['Year'].fillna(data['Year'].median())
data.isnull().sum()
```

```python
data['Kilometers_Driven'] = data['Kilometers_Driven'].fillna(data['Kilometers_Driven'].mean())
data.isnull().sum()
```

```python
cols = data[['Transmission', 'Owner_Type']]
for c in cols:
    data[c] = data[c].fillna(data[c].mode()[0])
    
data.isnull().sum()
```

```python
data.dtypes
```

```python
data['Fuel_Type'].nunique()
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

data['Fuel_Type'] = labelencoder.fit_transform(data['Fuel_Type'].values)


labelencoder1 = LabelEncoder()
data['Transmission'] = labelencoder1.fit_transform(data['Transmission'].values)
data
```

```python
data['Owner_Type'].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder 
onehotencoder = OneHotEncoder(drop='first')

owner_type_encoded = onehotencoder.fit_transform(data.iloc[:, 6:7].values).toarray()
owner_type_encoded = pd.DataFrame(data=owner_type_encoded, columns=['Owner Type Second', 'Owner Type Third'])
data.drop(columns=['Owner_Type'], inplace=True)

concatted_df = pd.concat([owner_type_encoded, data], axis=1)
concatted_df
```

```python
object_cols = data.select_dtypes(include=['object']).columns

for col in object_cols:
    mean_prices = concatted_df.groupby(col)['Price'].mean()
    overall_median = mean_prices.median()
    concatted_df['Categorized ' + col] =  concatted_df[col].map(lambda col: 1 if mean_prices[col] > overall_median else 0)
    concatted_df.drop(columns=[col], inplace=True)
    
concatted_df
```

```python
concatted_df.shape
```

```python
result = pd.DataFrame()
import numpy as np 
threshold = 3
outlier_indices = set()
    
for column in concatted_df.columns:
        
    mean = concatted_df[column].mean()
    std_dev = concatted_df[column].std()
        
    
    concatted_df[f'{column}_Z_Skoru'] = (concatted_df[column] - mean) / std_dev
        
    
    is_outlier = np.abs(concatted_df[f'{column}_Z_Skoru']) > threshold
    outlier_indices.update(concatted_df.index[is_outlier])
        
    concatted_df.drop(columns=[f'{column}_Z_Skoru'], inplace=True)
    
cleaned_df = concatted_df.drop(index=outlier_indices)
cleaned_df
```

```python
cleaned_df.corr()
```

```python
corrs = cleaned_df.corr()['Price'].drop('Price')
vals = corrs[abs(corrs) < 0.5].index.tolist()
cleaned_df.drop(vals, axis=1, inplace=True)
cleaned_df.head()
```

```python
x = cleaned_df.copy()
x.drop(columns='Price', inplace=True)
x.head()
```

```python
y = cleaned_df.copy()
y = cleaned_df.iloc[:, 4:5].values
y = pd.DataFrame(data=y, columns=['Price'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)

from sklearn.linear_model import LinearRegression
linear = LinearRegression()

linear.fit(x_train, y_train)

train_predict = linear.predict(x_train)
test_predict = linear.predict(x_test)

from sklearn.metrics import root_mean_squared_error, r2_score

rmse1 = root_mean_squared_error(y_train, train_predict)
rmse2 = root_mean_squared_error(y_test, test_predict)

train_r2 = r2_score(y_train, train_predict)
test_r2 = r2_score(y_test, test_predict)

print("TRAIN")
print(rmse1)
print(train_r2)

print("TEST")
print(rmse2)
print(test_r2)
```
